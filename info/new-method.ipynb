```python
# url variable store url 
url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json?location="+lat+","+lng+"&keyword=gluten_free&type=restaurant&language=es&rankby=distance&pagetoken&key="+APIKEY

print(url)
```

```python
# return response object 
r = requests.get(url) 

# json method of response object convert 
# json format data into python format data 
x = r.json() 
```

```python
data = pd.json_normalize(x['results'])
```

```python
if data.empty is True:
    url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json?location=" + lat + "," + lng + "&keyword=gluten_free&radius=3000&type=restaurant&language=es&key=" + APIKEY
    r = requests.get(url)
    x = r.json()
    data = pd.json_normalize(x['results'])
```
